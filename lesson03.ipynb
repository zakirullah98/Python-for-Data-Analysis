<h2> 3.1 Conditional Statements-条件语句 </h2>

<h3> If statement: </h3> 

* The **if-statement** is used to test a specific condition. If the condition is true, a block of code (if-block) will be executed.

```python
num = 9
if num%2 == 0: 
    # indentation
    print("Number is even")  
```

```python
# One line if statement:
if num%2 != 0: print("Number is not even")  
```

if statement example:-Program to print the largest of the three numbers.

```python
a = 10  
b = 9  
c = 11 

if a>b and a>c:  
    print("%s is largest"%(a)) 
if b>a and b>c:  
    print(f"{b} is largest") 
if c>a and c>b:  
    print("{} is largest".format(c))   
```

<h3>Nested If Statment</h3>

* `Nested If-Statement` means to place one `If-Statement` inside another `If-Statement`.

* There may be a situation when you want to check for another condition after a condition resolves to true. In such a situation, you can use the nested if construct.

```python
age = 19
isGraduated = False
hasLicense = True

# Look if person is 18 years or older
if age >= 18:
    print(f"You're {age} years old.")

    if isGraduated:
        print('Congratulations with your graduation!')
    if hasLicense:
        print('Happy driving!')
```

<h3> The if-else statement</h3>

* The **if-else statement** provides an **else** block combined with the **if-statement** which is executed in the false case of the **if** condition.

```python
#if-else example-1:

#Program to check whether a person is eligible to marry or not.

age = 21

if age>=22:  
    print("You are eligible to marry !!")
else:  
    print("Sorry! you have to wait !!") 
```

```python
#One line if-else statement:
age = 20

print("You are eligible to marry !!") if age>=22 else print("Sorry! you have to wait !!") 
```

```python
#if-else example-2:

#Program to check whether a number is even or not.

num = -8
if num%2 == 0:  
    print(f"{num} is even number.")  
else:  
    print(f"{num} is odd number.") 
```

<h3> Nested if  else-statment</h3>

* Nested `If else-Statement` means to place one `If else-statment` inside another `If else-Statement`.

```python
# program to detect weather a number is zero, possitive or negative
a = 0
if a <= 0:
    if a == 0:
        print(f"The given number is {a}")
    else:
        print("The given number is a Negative number")
else:
    print("The given number is a Positive number")
```

<h3> The elif statement </h3>

* The **elif-statement** adds another "decision" branch to **if-else**. 

* The **elif-statement** enables us to check several conditions and execute the specific block of statements depending upon the true condition among them. 



```python
#elif statement: Example-1

# Grades calculator

marks = 20

if marks > 85 and marks <= 100:    
   print("Congrats ! you scored grade A .")  

elif marks > 60 and marks <= 85:  
   print("You scored grade B + .")

elif marks > 40 and marks <= 60:  
   print("You scored grade B .") 

elif (marks > 30 and marks <= 40):  
   print("You scored grade C .")  

else:  
   print("Sorry you are fail!")  
```

<h3> Switch Case in Python</h3>

The if makes a decision based on whether the condition is true or not.  When we need to run several conditions, you can place as many elif conditions as necessary between the if condition and the else condition. Switch case statements are a substitute for long if statements that compare a variable to several integral values.

```python
day = 1

match day:
    case 1:
        print("Monday")
        break
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Not a weekday")
```

That’s the advantage Python’s native switch statement has over those of other languages. The break keyword's functionality is done for you behind the scenes.

<h2>3.2 Iterative Statments-循环语句</h2>

Python has two primitive loop commands:

1. while loops
2. for loops

<h3> While循环</h3>

* With the **while loop** we can execute a set of statements as long as a condition is true.


```python
# Python program to show how to use a while loop  
counter = 0  
# Initiating the loop  
while counter < 10: # giving the condition  
    counter = counter + 3  
    print(f"counter is  {counter}")
```

```python
# nested while loop
i = 1
while i <= 4 :
    j = 0
    while  j <= 3 :
        print(f"{i}*{j}=",i*j, end=" ")
        j += 1
    print(f",value of i={i}")
    print()
    i += 1
```

<h3>  Data structure construction using while loop</h3>

```python
# Python program to show how to use a while loop  
counter = 0  
# Initiating the loop 
dictionary={}
while counter < 10: # giving the condition  
    counter = counter + 1 
    dictionary[counter]=counter*counter
    print(f"counter is  {counter}")
print(f"The dictionary is created is: {dictionary}")
```

<h3>for循环</h3>

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

The for loop in python does not require an indexing variable to set beforehand.

```python
# for loop example
s="banana"
for x in s:
  print(x)
```

```python
# for loop example
for i in range(1, 11):
    print(i , end=' ')
    #print()
```

```python
# for loop example
l=[1,2,3,4,5,6,7,8]
for i in l:
    print(i * 2,end=' ')    
```

```python
# nested for loop
# outer loop
for i in range(1, 11):
    # nested loop
    # to iterate from 1 to 10
    for j in range(1, 11):
        # print multiplication
        print(i * j, end=' ')
    print()
```

```python
# Python program to show how the for loop works  
  
# Creating a sequence which is a list of numbers  
numbers = [4, 2, 6, 7, 3, 5, 8, 10, 6, 1, 9, 2]  
  
# variable to store the square of a number  
square = 0  
  
# Creating an empty dictionary  
squares = {}  
  
# Creating a for loop  
for value in numbers:  
    square = value ** 2  
    squares[f"square of {value}"] =square
print("The dictionary of numbers and its squares is:\n\n", squares) 
```

```python
string = "Python Loop"  
  
# Initiating a loop  
for c in  string:  
    # giving a condition in if block  
    if c == "o":  
        print(f"If block: {c}")  
    # if condition is not satisfied then else block will be executed  
    else:  
        print(f"else block: {c}")  
```

<h3>  Data structure Comprehension using for loop</h3>
Comprehensions in Python provide us with a short and concise way to construct new sequences (such as lists, set, dictionary etc.) using sequences which have been already defined. 

```python
# Using for-loop for constructing output list
given_list=[]
for x in range(5):
    given_list.append(x)
print(f"given_list:{given_list}")
```

```python
# list comprehension
given_list = [x for x in range(5)]
print(f"given_list:{given_list}")

new_list = [var+3 for var in given_list]

print(f"new_list:{new_list}")
```

```python
# list comprehension
given_list = [x for x in range(5)]
print(f"given_list: {given_list}")

# Using for-loop for constructing output dict
new_dict1={}
for x in given_list:
    new_dict1[x]=x+3

# dictionary comprehension
new_dict = {var:var + 3 for var in given_list }

print(f"new_dict:{new_dict}")
```

```python
# Python code to demonstrate dictionary  comprehension
 
# Lists to represent keys and values
keys = ['a','b','c','d','e']
values = [1,2,3,4,5] 
z=zip(keys,values)
print(list(z))
```

```python
# but this line shows dict comprehension here 
myDict = { k:v for (k,v) in zip(keys, values)}
myDict
```

```python
# We can use below too
myDict = dict(zip(keys, values)) 
```

<h2>  3.3 Transfer Statments </h2>

The transfer statement in python can be used to change the execution of the statement from its normal sequence order.These statements are often used in loops for and while. Let’s have a look at the transfer statements below.

<h3>continue语句</h3>

It returns the control to the beginning of the loop.

```python
# Python program to show how the continue statement works  
  
# Initiating the loop  
for c in "Python Loops":
    # case sensitive P!=p
    if c == "o" or c == "p" or c == "t":  
         continue  
    print('Current Letter:', c)  
```

<h3> Break Statement-中断语句</h3>

With the break statement we can stop the loop before it has looped through all the items.Break statement is put inside the loop body (generally after if condition).

```python
for i in range(10):
    print(i)
    if i == 2:
        break
```

```python
# Python program to show how the break statement works  
  
# Initiating the loop  
for c in "Python Loops":  
    if c == 'L':  
         break  
    print('Current Letter: ', c) 
```

<h3> Pass Statement-蟒蛇关</h3>

In Python programming, the pass statement is a null statement which can be used as a placeholder for future code.

```python
n = 12

# use pass inside if statement
if n > 10:
    pass

print('Hello')
```

```python
# Python program to show how the pass statement works  
for c in "Python Loops":  
    pass  
print( 'Last Letter:', c)   
```
