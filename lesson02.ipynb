The basic data structures in Python are.

<header><h2> Features of Data Structures/Collections</h2></header>   


* **Mutable Objects:**
Objects whose state can be changed once it is created like adding, updating or deleting elements(Lists, Dictionary, Set).
* **Immutable Objects:**
Once the object is created, we cannot add, remove or update elements.
String, Integer, Tuples are   immutable object types in python.

* **Homogeneous Data Structure:**
Data elements will be of same data type (ex: Array).
* **Heterogeneous Data Structure:** 
Data elements may not be of same data type (ex: List, Tuples, Sets etc).

* **Iterable:** an iterable is a collection of elements that you can loop (or iterate) through one element at a time (list, dict, tuple, and set etc.).

* **Ordered:**  An iterable is ordered  if you can retrieve its elements in a predictable order.

<h2> 2.2.4 Sets- </h2> 

In mathematics, a set is a well-defined collection of unique elements. 

The elements of a set are stored in an unorderly fashion meaning the items are randomly stored in the set and there is no definite position or index supported, neither slicing is allowed in a set. 

Elements can be added or removed from the set but cannot be changed as there is no concept of indexing and therefore elements cannot be changed. 

Set operations can be performed such as union, intersection, difference, disjoint. Lets look at how to implement it.

```python
# create an empty set
empty_set = set()
type(empty_set)
```

```python
s1 = {'a', 'b', 'c', 'c',1}
s2 = set({'a', 'd', 'e'})
```

```python
s1[0]
```

```python
# unordered
s1 = {'a', 'b', 'c', 'c',1}
s3 = {1, 'b', 'c', 'c','a'}
s3==s1
print(s3)
```

```python
s3[0]=30
```

```python
# unordered
s1.update(['m', 'n'])
s1
```

### Advantages of sets

### Set Comprehensions:
Set comprehensions are pretty similar to list comprehensions. The only difference between them is that set comprehensions use curly brackets { }. Lets look at the following example to understand set comprehensions.

```python
input_list = [1, 2, 3, 4, 4, 5, 6, 6, 6, 7, 7]
  
 
# Using set constructing output set
output_set={var for var in input_list if var % 2 == 0 }
  
print("Output Set using set comprehension:", output_set)
```

### Memory requirement (List vs Tuple vs Dict vs Set)
